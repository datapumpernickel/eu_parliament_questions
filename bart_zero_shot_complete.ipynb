```python
classifier = pipeline("zero-shot-classification",
                      model="facebook/bart-large-mnli", device = 0)

```

```python

candidate_labels = ['Agriculture',
'Banking',
"internal market",
'Civil Rights or Liberties or human rights',
'Culture or entertainment',
'Defence',
'Education',
'Energy',
'Environment or Climate',
'Employment',
'Foreign Affairs',
'Foreign Trade',
'Health',
'Immigration',
'Law and Crime',
'Macroeconomics',
'Regional Policy',
'Science, Technology and Communications',
'Social Policy',
'Transportation']

# think about 
# aid, merge internal market and macro economic or add 'taxes to it'
# maybe add gender
# take out social policy
```

```python
texts = pd.read_csv('https://box.fu-berlin.de/s/nnmFTYSsJ85zZbM/download').content.dropna().unique()
```

```python
complete_set = len(texts)
```

```python
while complete_set != len(final_results.sequence.unique()):
  final_results = pd.read_excel("drive/MyDrive/data_eu_questions/result.xlsx")
  sequences = pd.Series(texts).sample(25).unique()
  sequences = pd.Series(sequences)
  new_sequence = [sequence for sequence in sequences if sequence not in final_results.sequence.unique()]

  results = []
  for i in range(len(new_sequence)):
      result =  classifier(new_sequence[i], candidate_labels, multi_label=True)
      print(i)
      results.append(result)

  df = pd.concat([pd.DataFrame.from_dict(result) for result in results])
  g = df.groupby(["sequence"]).apply(lambda x: x.sort_values(["scores"], ascending = False)).reset_index(drop=True)
  g = g.groupby('sequence').head(3)
  full = g.append(final_results[["sequence","labels","scores"]])
  full.to_excel('drive/MyDrive/data_eu_questions/result.xlsx')
```
